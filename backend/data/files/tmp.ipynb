```python
import pandas as pd
cs = pd.read_csv("pickup.csv", index_col=0)
cs.head()
```

```python
st = cs.iloc[3, [0,5]].to_list()
st[1] = st[1].replace("[", "").replace("]", "").replace("'", "")
recipe = st[1].split(", ")
result = {
    'name': st[0],
    'recipe':[
        {
            'amount': recipe[i],
            'measure': 'cl',
            'name': recipe[i+1]
        } for i in range(0, len(recipe), 2)
    ]
}
print(result)
```
