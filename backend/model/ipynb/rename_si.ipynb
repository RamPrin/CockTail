## Fix Special Ingredients 
#### by renaming them to a standard name in recipes and add addition columns ```ingredients_list``` for the raw_ingredient with the special ingredient name

````python
type(df['raw_ingredients'][0][0][1])
````

````python
import re

def replace_special_ingredients(row):
    # Convert the recipe and raw_ingredients columns to string type
    
    # Find all special ingredients in the recipe
    special_ingredients = re.findall(r'Special ingredient #[0-9][^\n]*', str(row['recipe']))
    special_ingredients_sl = re.findall(r'special ingredient #[0-9][^\n]*', str(row['recipe']))
    special_Ingredients = re.findall(r'Special Ingredient #[0-9][^\n]*', str(row['recipe']))
    special_ingredients_no_number = re.findall(r'Special ingredient [A-Z][^\n]*', str(row['recipe']))
    special_ingredients_no = re.findall(r'Special ingredient No.[0-9]', str(row['recipe']))
    special_ingredients_num = re.findall(r'\*[0-9]', str(row['recipe']))
    special_ingredients_star = re.findall(r'\*\d+ [^\n]*', str(row['recipe']))
    special_ingredients_colon = re.findall(r'Special ingredient: [^\n]*', str(row['recipe']))
    special_ingredients.extend(special_Ingredients) 
    special_ingredients.extend(special_ingredients_no_number)
    special_ingredients.extend(special_ingredients_no)  
    special_ingredients.extend(special_ingredients_num)
    special_ingredients.extend(special_ingredients_star)
    special_ingredients.extend(special_ingredients_colon)
    special_ingredients.extend(special_ingredients_sl)
    
    # Replace the placeholders in the raw_ingredients column
    for ingredient in special_ingredients:
        # Extract the number of the special ingredient
        match = re.search(r'(\d+)', ingredient)
        if match:
            number = match.group(1)
        else:
            number = '1'  # Default to 1 if no number is found

        # Extract the name of the special ingredient
        name = re.search(r'\s?[=:-]? (.*?)(?= made as follows:| Ingredient #|\(|$)', ingredient)
        if name:
            name = name.group(1)
            # Remove 'ingredient #1 = ' from the name
            name = re.sub(r'(ingredient #\d+ = |Ingredient #\d+ - |Ingredient #\d+: |ingredient: )', '', name, flags=re.IGNORECASE)
            
            name = re.sub(r' made .*', '', name)
            name = re.sub(r' –.*', '', name)
            name = re.sub(r' - .*', '', name)
            name = re.sub(r': .*', '', name)
        else:
            continue

        # print(name)
        name = name.replace('* ', '').replace('*', '').replace(r'\d:\d|at \d:\d ratio', '').replace('4:3','')
        print(name)
        # Replace the placeholder with the actual name

        row['raw_ingredients'] = [[item[0], str(item[1]).replace(f'special ingredient #{number}', name)] for item in row['raw_ingredients']]
        row['raw_ingredients'] = [[item[0], str(item[1]).replace(f'{number}', name)] for item in row['raw_ingredients']]

    return row

df = df.apply(replace_special_ingredients, axis=1)
````

### The function to calculate the probabilities

````python
def calculate_cond_prob(main_ing, dep_ing):
    num_of_main = len(df[df[main_ing] != 0])
    num_of_dep = len(df[(df[main_ing] != 0) & (df[dep_ing] != 0)])
    if num_of_dep != 0:
        return num_of_dep/num_of_main
    else: return 0.0
````

````python
class Cocktail():
    def __init__(self) -> None:
        self.cocktail = []
        self.size = np.random.randint(3, 6)

    def new_ingredient(self, ingredient):
        self.cocktail.append(ingredient)
    
    def get_cur_size(self):
        return len(self.cocktail)
    
    def cocktail_components(self):
        return self.cocktail
    
    def is_finished(self):
        if len(self.cocktail) == self.size:
            return True
        else: return False

    def __str__(self):
        return f'{self.cocktail}'
    
    def __repr__(self) -> str:
        return self.__str__()
````

````python
class Cocktail_Generator():
    def __init__(self, table) -> None:
        self.prob_table = table
        self.cocktails = []
    
    def level_leaves(self, i):
        return max(4-(self.cocktails[i].get_cur_size()), 1)
    
    def main_ingredient(self, ingredient):
        for i in range(6):
            cocktail = Cocktail()
            cocktail.new_ingredient(ingredient)
            self.cocktails.append(cocktail)

    def check_value(self, eps):
        value = np.random.uniform(low=0.0, high=1.0)
        if value >= eps:
            return True
        else: return False
    
    def prob_row(self, branch):
        cur_prob = np.ones(shape=(1, len(self.prob_table)))
        for i in range((self.cocktails[branch]).get_cur_size()):
            ing = self.cocktails[branch].cocktail_components()[i]
            cur_prob = cur_prob * np.array(self.prob_table.loc[ing]).reshape(1, -1)

        return cur_prob[0]
    
    def select_next_ing(self, branch):
        probs = self.prob_row(branch)
        indices = np.argpartition(probs, -2)[-7:]
        collect_next_level = []
        i = -1
        leaves = 0
        level_leaves = self.level_leaves(branch)
        while leaves < level_leaves:
            if self.check_value(0.9):
                next_ing = np.random.choice(self.prob_table.columns)
            else:
                next_ing = self.prob_table.columns[indices[i]]
            if not next_ing in self.cocktails[branch].cocktail_components():
                collect_next_level.append(next_ing)
                leaves += 1
            i -= 1
        return collect_next_level
    
    def get_coctail(self, branch):
        return self.cocktails[branch].cocktail_components()
    
    def launch(self):
        next = self.select_next_ing(0)
        print(self.cocktails)
        print(next)
        for i in range(0,6,2):
                self.cocktails[i].new_ingredient(next[i//2])
                self.cocktails[i+1].new_ingredient(next[i//2])
                sec_next = self.select_next_ing(i)
                self.cocktails[i].new_ingredient(sec_next[0])
                self.cocktails[i+1].new_ingredient(sec_next[1])
        
        print(self.cocktails)
    
        for i in range(6):
            while not self.cocktails[i].is_finished():
                next = self.select_next_ing(i)[0]
                self.cocktails[i].new_ingredient(next)
        return self.cocktails
        
        
````

#### Create dataset that we need


````python
# "Melt" the correlation matrix
corr_df = corr_matrix.stack().reset_index()
corr_df.columns = ['ingredient1', 'ingredient2', 'correlation']

# Remove self-correlation entries
corr_df = corr_df[corr_df['ingredient1'] != corr_df['ingredient2']]
corr_df['ingredient1'] = corr_df['ingredient1'].astype(str)
corr_df['ingredient2'] = corr_df['ingredient2'].astype(str)

# Remove """ in the ingredient names in the correlation dataframe
corr_df['ingredient1'] = corr_df['ingredient1'].str.replace('"""', '').str.replace('"', '')
corr_df['ingredient2'] = corr_df['ingredient2'].str.replace('"""', '').str.replace('"', '')

# Sort by correlation
corr_df.sort_values(by='correlation', ascending=False, inplace=True)

# Remove duplicates
corr_df = corr_df[corr_df['ingredient1'] < corr_df['ingredient2']]

# Print the final dataframe
corr_df

# Save to a CSV file
corr_df.to_csv('data/correlation_data_.csv', index=False)
````
